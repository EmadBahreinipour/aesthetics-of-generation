# 01. Corpus Generation

This notebook demonstrates how to generate a corpus of texts from multiple LLMs.

## Overview
1. Load configuration and prompts
2. Initialize LLM providers
3. Generate texts across models
4. Save corpus for analysis

```python
import sys
sys.path.insert(0, '..')

from src.generation import LLMGenerator
from dotenv import load_dotenv
import pandas as pd

load_dotenv('../.env')
```

## 1. Load Configuration

```python
generator = LLMGenerator('../config.yaml')

generator_configs = generator.config['generation']
models = generator.config['models']

print("Models to compare:")
for model in models:
    print(f"  - {model['name']} ({model['provider']})")

print(f"\nGeneration settings:")
print(f"  temperature: {generator_configs['temperature']}")
print(f"  max_tokens:  {generator_configs['max_tokens']}")
print(f"  runs/prompt: {generator_configs['runs_per_prompt']}")
```

## 2. Register Providers

```python
generator.setup_providers()

print(f"Active providers: {list(generator.providers.keys())}")
```

## 3. Load Prompts

```python
prompts = generator.load_prompts('../data/prompts/prompts.yaml')

print(f"Loaded {len(prompts)} prompts across genres:")
genres = {}
for p in prompts:
    genres.setdefault(p['genre'], []).append(p['id'])
for genre, ids in genres.items():
    print(f"  {genre}: {len(ids)} prompts")
```

## 4. Generate Corpus

```python
corpus = generator.generate_corpus(
    prompts=prompts,
    models=['llama3'],
    output_dir='../data/raw',
)

successful = [r for r in corpus if 'error' not in r]
failed = [r for r in corpus if 'error' in r]
print(f"\nCorpus: {len(successful)} texts, {len(failed)} failures")
```
